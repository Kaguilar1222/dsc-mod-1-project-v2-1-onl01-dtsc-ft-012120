##  Importing our Data and Creating Tables

```python
#creating a list of the data we will be reading
csv_files = glob('./zippedData/*.csv.gz')
tsv_files = glob('./zippedData/*.tsv.gz')
files = csv_files + tsv_files
type(files)
```

```python
#cleaning file name
#creating dictionary of dataframes of each csv and tsv
files_dict = {}
for filename in files:
    if 'csv' in str(os.path.basename(filename)):
        filename_cleaned = os.path.basename(filename).replace(".csv.gz", "")\
        .replace(".", "_")
        filename_df = pd.read_csv(filename, compression='gzip', index_col=0)
        files_dict[filename_cleaned] = filename_df
    else:
        filename_cleaned = os.path.basename(filename).replace(".tsv.gz", "")\
        .replace(".", "_")
        filename_df = pd.read_csv\
        (filename, compression='gzip', delimiter ='\t', index_col=0)
        files_dict[filename_cleaned] = filename_df
```

```python
#viewing file names from k:v pairs in our dictionary
files_dict.keys()
```

```python
#checking a key in the dictionary to ensure that dataframes have been created 
files_dict['rt_movie_info'].info()
```

```python
#connecting to new database
conn = sqlite3.connect("movies_db.sqlite")
```

```python
#creating function to create a SQL table passing a dataframe, name, and 
#connection to database
def create_sql_table_from_df(df, name, conn):
    try:
        df.to_sql(name, conn)
        print(f"Created table {name}")
    except Exception as e:
        print(f"could not make table {name}")
        print(e)
```

```python
#passing the function through a for loop of the tables and names in our 
#dictionary of csv files
for name, table in files_dict.items():
    create_sql_table_from_df(table, name, conn)
```

```python
#creating a cursor to do a query of our sql db
cur = conn.cursor()
```

```python
#testing connection on our database of tables
conn.execute("SELECT name FROM sqlite_master WHERE type='table';").fetchall()
```

# Blockbuster Movie Season: Profits and Seasonality
Convention leads us to believe that profitable movies are released in the summer and holiday season (when, as is held, there is a higher volume of movie releases), so I suspect that the ranges of profit may be large in these movie-crowded months. 

## Question 1: Let's review the seasonality of movie releases by profit. What time of year would we want to release a high-budget, high-profit film?

### ETL for Profit/Release Dates
Let's clean the data and add relevant columns by determining the profit of each movie, the month of release, ensuring the data types are correct and removing duplicates.

```python
#creating a dataframe from the movie budgets file
budgets_and_release_dates_df = files_dict['tn_movie_budgets']
```

```python
#taking a look at the columns
budgets_and_release_dates_df.head()
```

```python
#reviewing the data types
budgets_and_release_dates_df.info()
```

```python
#changing datatype of release date to a datetime object to extract the month
budgets_and_release_dates_df['release_date'] = pd.to_datetime\
                                (budgets_and_release_dates_df['release_date'])
```

```python
#changing datatype of the budget and gross columns to int64
budgets_and_release_dates_df[budgets_and_release_dates_df.columns[2:]] \
        = budgets_and_release_dates_df[budgets_and_release_dates_df.columns[2:]]\
        .apply(lambda x: x.str.replace('$',''))\
        .apply(lambda x: x.str.replace(',','')).astype(np.int64)
```

```python
#creating profit column subtracting budget from worldwide gross
budgets_and_release_dates_df['movie_profit'] \
        = budgets_and_release_dates_df['worldwide_gross']\
        - budgets_and_release_dates_df['production_budget']
```

```python
#confirming effective datatype changes
budgets_and_release_dates_df.head()
```

```python
#creating month column so we can plot it against profit
budgets_and_release_dates_df['release_month'] = \
        pd.DatetimeIndex(budgets_and_release_dates_df['release_date']).month
```

```python
#checking to ensure no duplicate rows
duplicate_budgets_and_release_dates_df = \
        budgets_and_release_dates_df[budgets_and_release_dates_df.duplicated()]
```

### Data Review
I am reviewing the data through a SQL query to get a better sense of the scale of profit numbers.

```python
#creating SQL table to run a query of our data (not run in this instance since 
#table had been created in prior kernel and produces an error message)
budgets_and_release_dates_df.to_sql('budgets_release_dates', conn)
```

### Data Visualization
Let's plot our data using a boxplot to show the profits for each month. I am using a boxplot so we can see the IQR for movie profits and how they move from month to month. I am also removing the outliers but leaving in the whiskers to get a good sense of the full range of movie success by month. 

```python
#creating boxplot using release month and profit
x = budgets_and_release_dates_df['release_month']
y = budgets_and_release_dates_df['movie_profit']
f, ax = plt.subplots(figsize=(10,10))
sns.axes_style('whitegrid')
sns.set_context('talk') 
sns.boxplot(x, y, palette='Blues_d', showfliers=False)
plt.title('Time of Year for Movie Releases & Profit')
plt.ylabel('Profit in Billions')
plt.xlabel('Month of Movie Release') 
plt.show()
```

## Conclusion

### Recommendation

```python
#splitting out our genre column by commas to create list
non_en_movie_details_df['genres'] = non_en_movie_details_df['genres']\
                                    .apply(lambda x: x.split(","))
```
